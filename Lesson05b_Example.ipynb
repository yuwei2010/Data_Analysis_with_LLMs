```python
import datasurfer as ds
from datasurfer.lib_llm.utility import agent_tool
from datasurfer.lib_llm.tool import read_file, write_file, web_search, get_current_datetime, list_directory_contents, \
                                    parse_rss_feed, copy_file
from IPython.display import display, Markdown
```

```python
tools = [read_file, write_file, copy_file, list_directory_contents]
mcp_servers = 'http://localhost:8000/mcp'
```

```python
agent = ds.Agent('AAA', model='gpt-5-mini', tools=tools, mcp_servers=mcp_servers)
len(await agent.get_available_tools())
```

```python
agent = ds.Agent('LAA',  tools=tools, mcp_servers=mcp_servers)
print(agent.skills)
```

```python
Markdown(await agent.resp('Using "planning-with-files" as guidence to plan a web search for sport news from Stuttgart of the current month. '
'Return at least 5 news. Perform the task using "web-search" as gaidance. After that, write the results to a file called "stuttgart_sport_news.md"'))   
```
