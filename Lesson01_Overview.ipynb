## 四、如何使用大语言模型
### 1. Prompt Engineering

    - 提示（Prompt）是语言模型的输入。

    - 提示可以包含多模态数据，如文本和图像。

    - 提示至少应包含指导模型完成特定任务的文本指令。

    - 提示中应包含所有相关上下文信息。
    
        例如，要判断图片中是否有汽车，提示中必须包含该图片。

    - 提示中的指令应具体且明确输出格式。
    
        例如，要求模型输出“1”表示有车，“0”表示无车，以便后续统计。

    - 提示中可以包含示例，帮助语言模型理解任务。

少样本学习（Few-shot Learning）与零样本学习（Zero-shot Learning）的区别解释如下：


1. 少样本学习（Few-shot Learning）：
  
- 指的是模型在只有少量示例的情况下，仍能学习并完成新任务的能力。
  
- 通常会给模型提供几个相关的示例，让模型通过这些示例理解任务要求。
  
- 例如，给模型3个句子样例及其分类标签，然后让模型对新的句子进行分类。


2. 零样本学习（Zero-shot Learning）：
  
- 指模型在没有任何具体示例的条件下，直接理解并完成新任务的能力。
  
- 通过对任务的文字描述或指令，使模型能够立即执行任务，而不需要额外的训练示例。
  
- 例如，直接告诉模型“判断这句话的情感是积极还是消极”，模型根据指令完成任务，无需示例。


总结：

- 少样本学习需要少量示例辅助，零样本学习则完全依赖任务描述或指令。

- 两者都是解决样本不足情况下，模型泛化能力的重要方法。

### 2. Context Engineering
- 上下文工程（Context Engineering）致力于为大型语言模型（LLM）提供正确、相关且格式合适的信息，以完成任务，比提示工程范围更广。

- 优化上下文窗口，管理LLM输入区域，确保提供充足且定制的数据，类似于LLM的“内存”。

- 上下文工程涵盖整个信息提供系统，包括数据检索、处理与管理，而提示工程仅聚焦于单条文本指令。

- 适用于构建复杂AI代理和动态应用，处理多场景和错误恢复，支持系统自动学习和减少重复错误。

- 涉及多种技术和架构，如检索增强生成（RAG）、记忆系统、工具集成及多智能体系统，提升模型表现。

- 相较于提示工程，上下文工程是其演进和超集，更适合复杂多步骤任务和现实应用场景。

- 对提高AI系统准确性、效率和鲁棒性至关重要，是AI开发者日益重要的技能。

### 3. [API](https://ai.google.dev/gemini-api/docs?hl=de)

- 语言模型提供商如OpenAI通常提供网页接口，用户可以通过网页发送单个提示给模型。  

- 网页接口适合少量提示，但处理大量文本时不方便，因为需要逐个发送提示。  

- OpenAI的Python库能直接从Python发送提示给模型，并处理模型的回答，实现自动化数据加载、提示生成和后续处理。  

- Python库还可以将语言模型与其他工具集成，例如用语言模型编写数据处理代码并立即执行。
